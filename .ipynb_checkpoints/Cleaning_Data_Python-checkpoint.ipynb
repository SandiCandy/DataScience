## Schritt 1: Erstmal angucken

## Nächster Schritt: Fehlerhafte Daten
Die Spalte "age" sollte ausschließlich Zahlen im Bereich von 0 bis 120 enthalten.


```python
df['age']
```

Das betrifft den Eintrag 'old' und den Eintrag '-78'. Hier kann 
Wir erzeugen eine neue Spalte 'age_parsed', in der die Werte von object zu numeric konvertiert werden.

```python
df['age_parsed'] = pd.to_numeric(df['age'], errors='coerce')
df
```

Wir überprüfen im folgenden, welche Werte außerhalb des Wertebereichs liegen (Alter darf nur zwischen 0 und 120 liegen).

```python
df[(df['age_parsed'] < 0) | (df['age_parsed'] > 120)]
```

Das ist nur ein Eintrag. Hier kann man annehmen, dass es sich um einen Fehler handelt und Clair Skillern tatsächlich 78 ist. Das können wir also ändern.

```python
df.loc[df['age_parsed'] < 0, 'age_parsed'] *= -1
df
```

## Sonstiges: Redundate Daten
Die Spalte "fullname" enthält die Kombination der Spalten "firstname" und "lastname", ist also redundant. Diese kann also ohne Verlust entfernt werden. Außerdem können wir die Spalte age entfernen.

```python
df.drop(columns=['full_name', 'age'], inplace=True)
```
