Write a program to guess a number in between0 and 100!

```r
guess_number <- function() {
    print("Guess a number between 0 and 100 until you get it right.")
    x <- sample(0:100, 1)
    while (as.numeric(readline()) != x) {
        print("Next try.")
    }
    print("Hooray!")
}
```

## Plot-Function


```r
plot(esoph$agegp, esoph$ncases, xlab="Alter", ylab="Fälle", main="Esoph - Fälle pro Altersgruppe", col="blue")
```

Das gleiche nochmal über Boxplot... 

```r
boxplot (esoph$ncases ~ esoph$agegp, xlab="Alter", ylab="Fälle", main="Esoph - Fälle pro Altersgruppe", border="orange",lwd=3)
```

```r
plot(esoph$alcgp, esoph$ncases, xlab="Alkoholmenge", ylab="Fälle", main="Esoph - Fälle pro Alkoholmenge", col="blue")
```

## Histogramm

```r
hist(esoph$ncontrols, xlab="Nr of Cases", ylab="Häufigkeit", main="Esoph-Histogramm", col="red")
```

```r
boxplot (esoph$ncases ~ esoph$agegp, main="Esoph dataset",
         border="orange",lwd=3)
```

## Mehrere Plots zusammen


```r
# set the graphical display parameters to 3 rows and 2 columns
par(mfrow=c(3,2)) # mfrow adds plots per row, from left to right
# draw boxplots for cases and controls, per each group 
plot(esoph$ncases ~ esoph$agegp, xlab="agegp",border="red", lwd=1)
plot(esoph$ncontrols ~ esoph$agegp,xlab="agegp", border="blue", lwd=1)
plot(esoph$ncases ~ esoph$alcgp,border="red",xlab="alcgp", lwd=1)
plot(esoph$ncontrols ~ esoph$alcgp,border="blue", xlab="alcgp", lwd=1)
plot(esoph$ncases ~ esoph$tobgp,border="red",xlab="tobgp", lwd=1)
plot(esoph$ncontrols ~ esoph$tobgp,border="blue", xlab="tobgp", lwd=1)
# add a title outside of the plotting area
title("Boxplots of Cases (left) and Controls (right)",outer=TRUE, line=-2, cex.main=2)
```
